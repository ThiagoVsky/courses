```python
import _thread
import time

class Person:
    
    def __init__(self, nome, pontos, dano):
        self.nome = nome
        self.pontos = pontos
        self.dano = dano
        self.vencedor = False
    
    def ataque(self, oponente):
        oponente.pontos -= self.dano
        
    def perdeu(self, oponente):
        oponente.vencedor = True
        print("Venceu:", oponente.nome)
        print("Perdeu", self.nome)

pilha = []
luta = True

# Define a function for the thread
def print_time( threadName, delay, p):
    count = 0
    pilha.append(p)
    posicao = pilha.index(p)
    print("Iniciando", p.nome, "Posição", posicao)
    while True:
        
        if p.vencedor:
            print("Venceu", p.nome)
            break
        if p.pontos<=0:
            print("Perdeu", p.nome)
            break
        
        if len(pilha)==2 and posicao==0:
            p.ataque(pilha[1])
        elif len(pilha)==2 and posicao==1:
            p.ataque(pilha[0])
            
        time.sleep(delay)
        count += 1
        #p.pontos = p.pontos*0.80
        print("Pontos restantes do {}: {}".format(p.nome, p.pontos))
        if p.pontos <=0:
            if len(pilha)==2 and posicao==0:
                p.perdeu(pilha[1])
                break
            elif len(pilha)==2 and posicao==1:
                p.perdeu(pilha[0])
                break

# Create two threads as follows
try:
    _thread.start_new_thread( print_time, ("Thread-1", 3, Person("Gigante", 2000,100) ))
    #time.sleep(1)
    _thread.start_new_thread( print_time, ("Thread-2", 2, Person("Cavaleiro", 1000,300) ))
except:
    print ("Error: unable to start thread")
```
